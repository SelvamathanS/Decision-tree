# ***Decision Tree***

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
d=pd.read_csv("/content/diabetes.csv")
x=d.drop('Outcome',axis=1)
y=d['Outcome']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
# dc = DecisionTreeClassifier() #defaultly gini index
dc = DecisionTreeClassifier(criterion='entropy') # entropy
# dc = DecisionTreeClassifier(criterion='log_loss') #information gain
dc.fit(x_train,y_train)
```

```python
y_pred = dc.predict(x_test)
```

```python
acc = accuracy_score(y_test,y_pred)
acc
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
plot_tree(dc, filled=True, feature_names=x.columns, class_names=['No Diabetes', 'Diabetes'])
plt.title('Decision Tree Visualization')
plt.show()
```
